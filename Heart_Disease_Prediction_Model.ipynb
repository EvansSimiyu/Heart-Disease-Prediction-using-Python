***Importing the dependencies***

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

***Data Collection and Processing***

```python
heart_data = pd.read_csv("/content/drive/MyDrive/Heart Disease Prediction/heart_disease_data.csv")
#data summary
#first 5 rows
heart_data.head()
```

```python
#drop null and duplicate values from the dataset
heart_data.drop_duplicates(inplace = True)
heart_data.dropna(inplace = True)
heart_data
```

```python
#Splitting the data into training and testing sets
x = heart_data.drop(columns = 'target',axis = 1)
y = heart_data['target']
x_train,x_test,y_train,y_test =  train_test_split(x,y,stratify= y,test_size = 0.20)

```

***Data Standardization***

```python
scaler = StandardScaler()
scaler.fit(x,y)
```

***Training the Model***

Logistic Regression

```python
logis_model = LogisticRegression()
#training the model
logis_model.fit(x_train,y_train)
```

***Model Evaluation***

```python
#Accuracy Score
```

```python
#Evaluating accuracy on training data
x_training_prediction = logis_model.predict(x_train)
training_data_accuracy = accuracy_score(y_train,x_training_prediction)
print(training_data_accuracy)
```

```python
#Evaluating accuracy on testing data
x_testing_prediction = logis_model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test,x_testing_prediction)
print(testing_data_accuracy)
```

***Building a Predictive System***

```python
input_values = (42,0,2,120,209,0,1,173,0,0,1,0,2)
#changing input data to a numpy array
input_values_np = np.asarray(input_values)
input_vals_reshaped = input_values_np.reshape(1,-1)

#making prediction
target_prediction = logis_model.predict(input_vals_reshaped)
print(target_prediction)

if target_prediction == 1:
  print("The Patient's heart has  defect")
else:
  print("The patient's heart is Healthy")
```
